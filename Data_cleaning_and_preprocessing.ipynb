**STANDARDIZE TEXT COLUMNS**

```python
# Example: rename "No-show" column values from "Yes"/"No" to consistent casing
if 'No-show' in data.columns:
    data['No-show'] = data['No-show'].str.strip().str.lower()
    data['No-show'] = data['No-show'].replace({'yes':'Yes', 'no':'No'})
```

```python
data.dtypes
```

**CONVERT AND CLEAN DATA TYPES**

```python
# Convert appointment date and scheduled day to datetime
if 'AppointmentDay' in data.columns:
    data['AppointmentDay'] = pd.to_datetime(data['AppointmentDay'], errors='coerce')
if 'ScheduledDay' in data.columns:
    data['ScheduledDay'] = pd.to_datetime(data['ScheduledDay'], errors='coerce')
```

```python
data.dtypes
```

```python
# Convert numeric-like columns to int
for col in ['Age', 'Scholarship', 'Hipertension', 'Diabetes', 'Alcoholism', 'Handcap', 'SMS_received']:
    if col in data.columns:
        data[col] = pd.to_numeric(data[col], errors='coerce').fillna(0).astype(int)
```

**RENAME COLUMNS (CLEAN FORMAT)**

```python
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('-', '_')
```
